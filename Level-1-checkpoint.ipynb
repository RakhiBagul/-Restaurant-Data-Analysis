# Task-1 : Top Cuisines

Determine the top three most common cuisines in the dataset.

```python
cuisines = df['Cuisines'].dropna().str.split(',').explode().str.strip()
Top_3_Cuisines = cuisines.value_counts().head(3)
print(Top_3_Cuisines)
```

```python
sns.barplot(x=Top_3_Cuisines.index, y=Top_3_Cuisines.values)
plt.title("Top 3 Common Cuisines")
plt.xlabel("Cuisines")
plt.ylabel("Count")
plt.show()
```

Calculate the percentage of restaurants that serve each of the top cuisines

```python
cuisines_in_percentages = (Top_3_Cuisines/len(df))*100
print(cuisines_in_percentages)
```

```python
sns.barplot(x=cuisines_in_percentages.index, y=Top_3_Cuisines.values, palette = "plasma")
plt.title("Top 3 Common Cuisines")
plt.xlabel("Cuisines")
plt.ylabel("cuisines_in_percentages")
plt.show()
```

# Task-2 : City Analysis

Identify the city with the highest number of restaurants in the dataset

```python
city_counts = df['City'].value_counts()
top_city = city_counts.idxmax()
top_city_count = city_counts.max()
print(top_city)
print(f"The city with the highest number of restaurants is {top_city} with {top_city_count} restaurants.")
```

Calculate the average rating for restaurants in each city

```python
city_avg_rating = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)
print(city_avg_rating)
```

```python
city_avg_rating = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(5)
print(city_avg_rating)
```

 Determine the city with the highest average rating

```python
city_avg_rating = df.groupby('City')['Aggregate rating'].mean()
top_rated_city = city_avg_rating.idxmax()
highest_avg_rating = city_avg_rating.max()
print(top_rated_city)
print(f"The city with the highest average rating is {top_rated_city} with an average rating of {highest_avg_rating:.2f}.")
```

# Task-3 : Price Range Distribution

Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

```python
price_range_counts = df['Price range'].value_counts().sort_index()
plt.figure(figsize=(8,5))
sns.barplot(x=price_range_counts.index, y=price_range_counts.values, palette="pastel")

plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Price Ranges Among Restaurants")
plt.xticks(rotation=0)  
plt.show()
```

 Calculate the percentage of restaurants in each price range category.

```python
price_range_counts = df['Price range'].value_counts()
price_range_percentage = (price_range_counts / df.shape[0]) * 100
print(price_range_percentage)
```

# Task -4 : Online Delivery

Determine the percentage of restaurants that offer online delivery.

```python
online_delivery_percentage = (df['Has Online delivery'].value_counts(normalize = True)*100).round(2)
print("Percentage of restaurants offering online delivery:")
print(online_delivery_percentage)
```

```python
plt.figure(figsize=(6, 6))
online_delivery_percentage.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90)
plt.show()
```

Compare the average ratings of restaurants with and without online delivery.

```python
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
avg_ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean().round(2)
print("Average Ratings of restaurants :")
print(avg_ratings)
```
